```python
d1=data.groupby('CustomerID')['Itemname'].apply(list).to_list()
```

```python
d1[8]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
pre=TransactionEncoder()
```

```python
res1=pre.fit_transform(d1)
```

```python
d2=pd.DataFrame(res1,columns=pre.columns_)
```

```python
d2.head()
```

```python
d2.replace(False,0,inplace=True)
d2.replace(True,1,inplace=True)
```

```python
d2.head()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
support=apriori(d2,min_support=.05,use_colnames=True)
```

```python
support.sort_values(by='support',ascending=False).head()
```

```python
confidence=association_rules(support,metric='confidence',min_threshold=.5)
```

```python
confidence.sort_values(by='confidence',ascending=False).head()
```

lift shows correlation between items ,it can be +ve or -ve ,basically it shows strength of association

```python
final=confidence[confidence['lift']>1] #posetive correlation 
```

```python
final.sort_values(by='lift',ascending=False).head()
```

```python
final2=confidence[confidence['lift']<1] #negetive correlation 
```

```python
final2.sort_values(by='lift',ascending=False).head()
```

```python
final3=confidence[confidence['lift']==1] #no correalation
```

```python
final3.sort_values(by='lift',ascending=False).head()
```
